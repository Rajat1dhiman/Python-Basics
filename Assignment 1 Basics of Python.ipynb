12. How does the elif statement work?

Ans. The elif(short for"else if") statement in Python used to check multiple conditions , allowing you to handle more than two cases in a condition block. It follows an **if** statement and can be used to check additional condition if the initial **if** condition evaluates to **False**.

**Working**
1. The **if** statement is evaluated to **true** than it will executed in the block.
2. The **if** statement is evaluate to **false** than it will move to other elif statements.
3. The **elif** statement allows you to specify additional conditions, which will be checked in order.

```python
age=21
if age>13:
  print("You are child.")
elif age<18:
  print("You are teenager")
elif age<30:
  print("You are young adult")
else:
  print("You are adult")
```

13. What is the difference between for and while loops?

ans.1. **for loop**
1. Purpose: Used to iterate over a sequence (like a list, tuple etc) or other iterable objects.
2. Structure: The loop runs a specified number of times as it iterates over elements in the sequence

```python
my_list=[1,2,3,4,5]
for item in my_list:
  print(item)
```

2. **while loop**
    1. Purpose: Repeated a block of code as long as a specified condition is **true**
    2. Structure: The loop runs until the condition becomes false.

```python
count=0
while count<5:
  print(count)
  count+=1
```

14. Describe a scenario where a while loop is more suitable than a for loop.

Ans. A scenario where while loop is more suitable than for loop is when the number of iterations cannot be pre- determined and depends on a condition that is checked during the execution of loop.

**Example:Password validation**

```python
# define the valid password
valid_password="Admin@987"
#start  while loop
while True:
  user_input= input ("Enter Your Password")
  if user_input==valid_password:
    print("Access Granted.")
    break# exit the loop when the correct password is entered
  else:
    print("Invalid Password, Please try again.")
```

10. Write a code to demonstrate type casting with list elements.

```python
# Original list with heterogeneous data elements
my_list=[1,2,3.5,"Banana",True]
# print original types of elements
print("Original list elements and their types:")
for element in my_list:
  print(f"{element}:{type(element)}")

```

```python
# type casting elements
casted_list=[int(element)if isinstance(element,float)else str(element)for element in my_list]
print("Casted list elements and their types:")
for element in casted_list:
  print(f"{element}:{type(element)}")
```

14. Write a program to reverse a string using  a while loop

```python
input_string=input("Enter a string:")
reversed_string=""
index = len(input_string) - 1
while index >= 0:
  reversed_string += input_string[index]
  index -= 1
print(f"Reversed string:{reversed_string}")
```

15. Write a Python program to calculate the factorial of number provided by user using a while loop

```python
# Take input from the user
num=int(input("Enter a number"))
# here we use factorial
factorial=1
# check if the number is positive ,negative
if num<0:
  print("Factorial is not defined for negative numbers")
else:
  while num>1:
    factorial *= num
    num -= 1
print(f"The factorial is:{factorial}")
```
